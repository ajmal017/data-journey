Before we do a correlation analysis, we can use corelation table and heatmap. In this way, we can have little knowledge of what variables we will compare.

```python
import matplotlib.pyplot as plt
f,ax = plt.subplots(figsize = (18, 18))
sns.heatmap(data.corr(), annot =True, linewidths =.5,fmt = '.1f',ax=ax);
```

According to graph, there is a very strong positive relationship between EU_Sales and Global_Sales variables. We can also say same thing about between NA_Sales and Global_Sales variables.

```python
na = data['NA_Sales']
eu = data['EU_Sales']
gb = data['Global_Sales']
```

```python
plt.scatter(na,gb,c='r',alpha = 0.5)
plt.xlabel('North America Sales')
plt.ylabel('Global Sales')
plt.title('Norh America Sales - Global Sales Correlation Chart');
```

```python
plt.scatter(eu,gb,c='g',alpha = 0.5)
plt.xlabel('Europe Sales')
plt.ylabel('Global Sales')
plt.title('Europe Sales - Global Sales Correlation Chart');
```

We can say that there is a direct proportion between Europe Sales/NA Sales and Global Sales. It means if the game is sold in Europe or Norh America, it is sold all over the world.

# Homework2 - Data Science Toolbox

```python
def threetimes (x):
    return x*3
a = threetimes(2)
print(a)
    
```

```python
def func1():
    def func2():
        a = 2
        b = 8
        return b/a
    return func2()/2
print(func1())
```

```python
def func3(x, y = 1,z = 1):
    return x*(y+z)
print(func3(2))
print(func3(2,2,2))
```

```python
def func4(*args):
    for i in args:
        print(i)
func4(1)
func4(5,6,1,4)
```

```python
def func5(**kwargs):
    for k,v in kwargs.items():
        print(k,':', v)
func5(Movie = 'Avatar', Year = 2009, Genre = 'Fantasy')
```

```python
threetimes = lambda x: x*3
print(threetimes(2))
```

```python
import numpy as np
vec = np.arange(1,10,1)
y = map(lambda x: x**2,vec)
print(list(y))
```

```python
list1 = list(np.arange(1,5))
list2 = list(np.arange(5,9))
zlist = zip(list1,list2)
print(zlist)
z_list = list(zlist)
print(z_list)
```

```python
unzip = zip(*z_list)
unlist1,unlist2 = list(unzip)
print(unlist1)
print(unlist2)
```

```python
x1 = [1,2,3]
x2 = [ i + 1 for i in x1]
print(x2)
```
